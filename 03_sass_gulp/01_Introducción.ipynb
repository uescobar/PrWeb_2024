## Dependencias del proyecto

Existen dependencias globales y dependencias del proyecto

### Dependencias globales

El comando `npm -v` hace uso de una dependencia global, no importa en qué terminal se ejecute.

Las dependencias del proyecto son específicas para cada uno de los proyactos y no tienen asociado un comando sino que existen en el archivo **package.json**

Para el caso que nos ocupa, _gulp va a ser una dependencia global._

Instalar gulp vía global, usando la terminal (Powershell o bash) se realiza usando el comando con la opción `g`, se especifica que es una dependencia global

`npm i -g gulp-cli`

y luego verificar su instalación con 

`gulp -v`

Nota para Windows: Si los comandos previos marcan error (letras rojas) hay que ejecutar los siguientes comandos desde la consola de PowerShell con permisos de administrador.

`Get-ExecutionPolicy`

`Set-ExecutionPolicy Unrestricted`, usar una letra O mayuscula para indicar que sí a todo.




Despues de eso, volver a probar el comando

`gulp -v`

### Dependencias locales

Para instalar una dependencia local, desde la terminal de VisualStudio Code

`npm i gulp`

Lo va a instalar y se agrega al archivo **package.json**, y ese archivo se puede compartir con un grupo de trabajo.

El archivo **package-lock.json** instala las dependencias de las dependencias, este archivo nunca se debe de modificar y la carpeta de node modules va a tener todas las dependencias de todo lo que se va instalando y es importante mantenerla, al menos en lo que se va desarrollando el proyecto.

Si la carpeta se borra, se pueden reinstalar las dependencias ejecutando `npm i` en la consola del proyecto.

### Dependencias y dependencias de desarrollo 

sass es una dependencia de desarrollo porque se sube a produccion la version compilada

Para instalar una dependencia de desarrollo se hace con el comando

`node i --save-dev gulp gulp-sass sass`

Cuando instala gulp lo instala en una categoría especial de desarrollo que se puede ver al final del JSON.


## Trabajando con GULP y SASS

GULP cuenta con una serie de apis muy interesantes que se pueden apreciar en el sito web oficial de [gulp](http://gulpjs.com) en la seccion de apis. Gulp es una api para comenzar a escribir tareas pero se iran agregando apis segun sea necesario

Para comenzar a trabajar con gulp hay que crear el archivo **gulpfile.js** y allí es donde se van a buscar las tareas.

El contenido del archivo **gulpfile.js** es:

**gulpfile.js**
***
```js
function compilarSASS(){
    console.log('Complilando...');
}

exports.compilarSASS = compilarSASS

```
***

Luego escribir en la consola el comando

`gulp`

y luego

`gulp --tasks`

`gulp compilarSASS`


No obstante, gulp sigue ciertas reglas, por lo que es mejor llevar un orden y crear la estructura de directorios adecuada para poder trabajar de una mejor manera

Crear la estructura de directorios al mismo nivel que **package.json**
```
src/
src/js
src/scss
src/scss/app.scss  
```
En el archivo **app.scss** vamos a escribir todo el codigo de sass, pero hay que hacer unas modificaciones al archivo gulpfile.js para decirle en donde está app.scss, quedando el contenido de la siguiente manera

**gulpfile.js**
***
```js
const { src } = require('gulp');

function compilarSASS(){
    // console.log('Complilando...');
    return src("./src/scss/app.scss")
}

exports.compilarSASS = compilarSASS

```
***

después de editar el archivo y a guardarlo, ejecutar en la terminal

`gulp compilarSASS`

Una vez que se vea el resultado del comando anterior, editar el archivo para dejarlo como sigue, 
en la primera linea el require accede a la libreria de gulp pero solo usa las funciones de src, dest,
el uso de require en este caso se usa como una instruccion para cargar la librería:

**gulpfile.js**
***
```js
const { src, dest } = require('gulp')
// const sass = require('gulp-sass')
const sass = require('gulp-sass')(require('sass'))



function compilarSASS(){
    // console.log('Complilando...');
    return src("./src/scss/app.scss")
        .pipe( sass() )
        .pipe( dest("./build/css") )
}

exports.compilarSASS = compilarSASS

```
***

Posteriormente a la edición del código, ejecutar en la terminal

`gulp compilarSASS`


Ahora es momento de editar el archivo **app.scss** y escribir:
**app.scss**
***
```scss
$rojo : 'red';

body{
    background-color: $rojo;
}
```
***

El código anterior representa el manejo de variables dentro del archivo scss y usando gulp, este será compilado y el resultado de la compilación será colocado dentro del directorio **./build/css** dentro del proyecto.

Las tareas que deseamos automaticar pueden ir encadenandose por medio de la sentencia **pipe**.

La ultima version del archivo **gulpfile.js** es:

**gulpfile.js**
***
```js
const { series, src, dest } = require('gulp');

const sass = require('gulp-sass')(require('sass'))

// Funcion que compila SASS

function css(  ){
   return src("./src/scss/app.scss")
    .pipe( sass({
        outputStyle: 'compressed'
    }) )
    .pipe( dest('./build/css') )
}

exports.css = css;

```
***

La opción **outputStyle** es opcional, y lo que realiza es que comprime el código (lo hace no muy legible para los seres humanos) de tal forma que no haya tanto formato al interior del archivo.
