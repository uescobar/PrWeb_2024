# Creando una tabla


MySQL tiene más de 40 tipos de datos diferentes, muy pocas veces se ocuparán todos.

Los tipos de datos más comúnes son:

* Números (Enteros)
        | TIPO | RANGO | UNSIGNED |
        | INT  | -2100 millones a 2100 millones | 0 a 4200+ millones | 
        | TINYINT | -128 a 127 | 0 a 255 |
        | SMALLINT | -32768 a 32767 | 0 a 65535 |
        | MEDIUMINT | -8.3 millones a 8.3 millones | 0 a 16.7 millones |
        | BIGINT | -2⁶³ A 2⁶³-1 | 0 a 2⁶⁴-1 |
        
* Números (Decimales)
        | TIPO | USOS |
        | FLOAT | Valores decimales |
        | DOUBLE | Valores decimales muy grandes |
        | DECIMAL | Valores donde no se aceptan redondeos (dinero) |
        


* Texto y datos binarios

        |  TIPO | USOS |
        | VARCHAR | Texto corto o texto que varía en su extensión (hasta 255 caracteres) |
        | CHAR | Corto, extensión fija (como un password encriptado) |
        | TEXT | Texto largo (como una entrada de blog |
        | ENUM | Un valor de una lista enumerada |
        | BLOB | Imágenes, sonidos, y archivos comprimidos |

* Fechas y Horas 
 
        | TIPO | USOS |
        | DATE | AAAA:MM:DD |
        | TIME | hh:mm:ss |
        | DATETIME | AAAA-MM-DD hh:mm:ss |
        | TIMESTAMP | AAAA-MM-DD hh:mm:ss |
        | YEAR | AAAA |

Para crear una tabla dentro de una base de datos usaremos el comando

`CREATE TABLE`

De preferencia, los nombres de las tablas deben escribirse en minúsculas. 


Ahora, la instrucción a ejecutarse  para crear la tabla servicios es:

```
CREATE TABLE servicios (
id INT(11) NOT NULL AUTO_INCREMENT,
nombre VARCHAR(60) NOT NULL,
precio DECIMAL(6,2) NOT NULL,
PRIMARY KEY (id)
);

```

***
```
MariaDB [(none)]>  USE appsalon;
Database changed

MariaDB [appsalon]> CREATE TABLE servicios (
    -> id INT(11) NOT NULL AUTO_INCREMENT,
    -> nombre VARCHAR(60) NOT NULL,
    -> precio DECIMAL(6,2) NOT NULL,
    -> PRIMARY KEY (id)
    -> );
    
Query OK, 0 rows affected (0.067 sec)

MariaDB [appsalon]> 

```
***


Ahora verificamos que se haya creado la tabla con su estructura

***
