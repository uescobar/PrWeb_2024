### Funciones

Repasaremos un poco mas JavaScript, llamando una función desde un boton en un formulario.


***
```HTML
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>JavaScript Demo</title>
    <script language="JavaScript" type="text/JavaScript">
        function sayHello() {
            document.write("Hola Mundo")
        }
    </script>
</head>
<body>
    <p>Dar click en el siguiente botón para llamar la función </p>
    <form>
        <input type="button" onclick="sayHello()" value="decir Hola Mundo">
    </form>
</body>
</html>

```

### Eventos


***
```HTML
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Demo 2</title>

    <script language="JavaScript" type="text/JavaScript">
        function mouseOver(){
            document.images["img1"].src = "logo_red.png"
        }

        function mouseOut() {
            document.images["img1"].src = "logo_green.png"
        }

        function load(){
            document.images["img1"].src = "logo_green.png"
        }

    </script>
</head>
<body onload="load()">
    <img name="img1" onmouseover="mouseOver()" onmouseout="mouseOut()">
</body>
</html>
```

### Manejo de errores

Se pueden manejar 

* Errores de tiempo de ejecución
* Se llaman excepciones
* Errores lógicos o de sintáxis
* se usan **try**, **catch**, **fynally** y **throw**


***
```HTML
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Demo 3</title>
    <script language="JavaScript" type="text/JavaScript">
        function myFunc() {
            
            let a = 100;

            try {
                let s  = a.toUpperCase();
                alert("El valor de la variable es: " + a);
            } catch (e) {
                alert("Error");
            } finally {
                alert("Este bloque siempre será ejecutado")
            }
        }
    </script>
</head>
<body onload="myFunc()">
    
</body>
</html>
```

### Cookies

Se usan para

* Mantener el estado en el lado del cliente.
* Las cookies son almacenadas en la computadora del cliente como un archivo de texto, su ubicacióp varía de navegador en navegador.
* Cuando los usuarios realizan una visita al sitio nuevamente, el sitio lee los datos desde la cookie.

Existen dos tipos de Cookies

* Session Cookie, solo esta disponible para sesión que se apertura en la página web
* Persistent Cookie, se almacena en la computadora y debe ser removida manualmente

***
```HTML
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Demo 4</title>
    <script language="JavaScript" type="text/JavaScript">
        function writeCookie(){
            if(document.myform.cliente.value == ""){
                alert("Ingresa algún valor:");
                return;
            }
            cookievalue = escape(document.myform.cliente.value) + ":";
            document.cookie = "nombre" + cookievalue;
            document.write("Registrando cookies : " + "nombre = " + cookievalue);
        }
    </script>
</head>
<body>
    <form name="myform" action="">
        Ingrese el nombre: <input type="text" name="cliente">
        <input type="button" value="Establecer cookie" onclick="writeCookie();">
    </form>
</body>
</html>
```

### Animation

***
```HTML
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=<device-width>, initial-scale=1.0">
    <title>Document</title>
    <script>
        let arr = new Array("1.png", "2.png", "3.png", "4.png");
        let i = 1;
        let ref;

        function picLibrary(){
            document.images[0].src = arr[i];
            i++;
            if (i > 3)
                i = 0;
        }

        function start(){
            ref = setInterval("picLibrary()", 1000);
        }

        function stop(){
            clearInterval(ref);
        }
    </script>
</head>
<body>
    <img src="1.png" name="img1" onmouseover="stop()" onmouseout="start()"/>
</body>
</html>
```

### Page Redirect


***
```HTML
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Demo 6</title>
    <script language="JavaScript" type="text/JavaScript">
        function redirect() {
            window.location="http://www.upt.edu.mx";
        }
    </script>
</head>
<body>
    <p>Dar click en el siguiente botón para ser redirigido a la pagina de la UPT</p>
    <form>
        <input type="button" value="Redirigeme" onclick="redirect();" />
    </form>
</body>
</html>
```
