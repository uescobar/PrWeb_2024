# Diagramas de Entidad Relación (E-R)

* Muestran de forma gráfica las Entidades (Tablas) y sus atributos
* Ayuda a conocer la forma en la que se relacionan los datos
* Existen muchas herramientas para generar estos diagramas

![01](08_img/01.png)

Las líneas tienen un significado y se le conoce como **cardinalidad**.

La **cardinalidad** se refiere al número máximo de veces que una instancia se relaciona con otra.

**No se debe pensar en esto como un número obligatorio, sino como el máximo de veces que 2 tablas se pueden relacionar**.

![02](08_img/02.png)

En la primer imagen, podemos deducir que entre la tabla de clientes y citas la cardinalidad es de uno a muchos (un cliente puede tener muchas citas).

La relación se establece por medio de la Llave foránea (FK, que significa Foreing Key)
