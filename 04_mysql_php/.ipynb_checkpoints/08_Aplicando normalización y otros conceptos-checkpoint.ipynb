# Diagramas de Entidad Relación (E-R)

* Muestran de forma gráfica las Entidades (Tablas) y sus atributos
* Ayuda a conocer la forma en la que se relacionan los datos
* Existen muchas herramientas para generar estos diagramas

![01](08_img/01.png)

Las líneas tienen un significado y se le conoce como **cardinalidad**.

La **cardinalidad** se refiere al número máximo de veces que una instancia se relaciona con otra.

**No se debe pensar en esto como un número obligatorio, sino como el máximo de veces que 2 tablas se pueden relacionar**.

Los significados de las relaciones son:

![02](08_img/02.png)

En la primer imagen, podemos deducir que entre la tabla de clientes y citas la cardinalidad es de uno a muchos (un cliente puede tener muchas citas).

La relación se establece por medio de la Llave foránea (FK, que significa Foreing Key), **una llave foránea es una llave primaria de otra tabla**

Ahora, la base de datos que tenemos, tiene que tener el siguiente diseño:

![03](08_img/03.png)

# Crear tabla pivote

Va a contener id's de la tabla **citas** y de la tabla **servicios** por eso se le llama tabla pivote, usualmente esta tabla pivote tiene un nombre compuesto **citasservicios**, que usualmente es el nombre de las dos tablas que va a unir. 

Crear la tabla **citasServicios**

`CREATE TABLE citasServicios (
     id INT(11) AUTO_INCREMENT,
     citaId INT(11) NOT NULL,
     servicioId INT(11) NOT NULL,
     PRIMARY KEY (id),
     KEY citaId (citaId),
     CONSTRAINT citas_FK
     FOREIGN KEY (citaId)
     REFERENCES citas (id),
     KEY servicioId (servicioId),
     CONSTRAINT servicios_FK
     FOREIGN KEY (servicioId)
     REFERENCES servicios (id)
     );
`

***
```
mysql> CREATE TABLE citasServicios (
    ->      id INT(11) AUTO_INCREMENT,
    ->      citaId INT(11) NOT NULL,
    ->      servicioId INT(11) NOT NULL,
    ->      PRIMARY KEY (id),
    ->      KEY citaId (citaId),
    ->      CONSTRAINT citas_FK
    ->      FOREIGN KEY (citaId)
    ->      REFERENCES citas (id),
    ->      KEY servicioId (servicioId),
    ->      CONSTRAINT servicios_FK
    ->      FOREIGN KEY (servicioId)
    ->      REFERENCES servicios (id)
    ->      );
Query OK, 0 rows affected (0.02 sec)
```

## Múltiples joins y más sobre BD

En la consulta anterior, hacen falta los datos del cliente.

`SELECT * FROM citasServicios
    LEFT JOIN citas ON citas.id = citasServicios.citaId
    LEFT JOIN clientes ON citas.clienteId = clientes.id
    LEFT JOIN servicios ON servicios.id = citasServicios.servicioId;
