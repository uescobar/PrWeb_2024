### MySQLi API

## Create Table

**mysqli_create.php**

***
```PHP
    <?php
// Configuración de la conexión a la base de datos
$servername = "localhost";
$username = "usuario";
$password = "contraseña";
$dbname = "testdb";

// Crear una conexión a la base de datos
$conn = new mysqli($servername, $username, $password, $dbname);

// Verificar la conexión
if ($conn->connect_error) {
    die("Error de conexión: " . $conn->connect_error);
}

// Consulta SQL para crear la tabla "employee"
$sql_create_table = 'CREATE TABLE employee( '.
                    'emp_id INT NOT NULL AUTO_INCREMENT, '.
                    'emp_name VARCHAR(20) NOT NULL,'.
                    'emp_address VARCHAR(20) NOT NULL, '. 
                    'emp_salary INT NOT NULL, '.
                    'PRIMARY KEY (emp_id))';

// Ejecutar la consulta para crear la tabla
if ($conn->query($sql_create_table) === TRUE) {
    echo "Tabla 'employee' creada exitosamente.<br>";

    // Consulta SQL para agregar un registro a la tabla
    $sql_insert_record = "INSERT INTO employee (emp_name, emp_address, emp_salary) VALUES ('Juan Perez', 'Calle Principal', 50000)";

    // Ejecutar la consulta para agregar el registro
    if ($conn->query($sql_insert_record) === TRUE) {
        echo "Registro agregado exitosamente.";
    } else {
        echo "Error al agregar el registro: " . $conn->error;
    }
} else {
    echo "Error al crear la tabla: " . $conn->error;
}

// Cerrar la conexión
$conn->close();
?>
    

```

## Generar reporte PDF

Para generar un reporte en PDF de la información en la tabla "employee", puedes utilizar una librería PHP como TCPDF o FPDF. En este ejemplo, te mostraré cómo hacerlo utilizando TCPDF, que es una librería popular y fácil de usar para la generación de PDF en PHP.

Primero, asegúrate de haber descargado TCPDF desde su sitio web oficial o utilizando Composer si estás trabajando en un proyecto PHP que utiliza Composer. Luego, crea un archivo PHP llamado generar_reporte.php con el siguiente código:

**generar_reporte.php**


***
```PHP
<?php
// Incluir la librería TCPDF
require_once('tcpdf/tcpdf.php');

// Configuración de la conexión a la base de datos
$servername = "localhost";
$username = "usuario";
$password = "contraseña";
$dbname = "testdb";

// Crear una conexión a la base de datos
$conn = new mysqli($servername, $username, $password, $dbname);

// Verificar la conexión
if ($conn->connect_error) {
    die("Error de conexión: " . $conn->connect_error);
}

// Consulta SQL para obtener los datos de la tabla "employee"
$sql_select_data = "SELECT * FROM employee";
$result = $conn->query($sql_select_data);

// Crear un nuevo objeto TCPDF
$pdf = new TCPDF('P', 'mm', 'A4', true, 'UTF-8', false);

// Establecer el título del documento
$pdf->SetTitle('Reporte de Empleados');

// Agregar una página al documento
$pdf->AddPage();

// Configurar estilo y formato del texto
$pdf->SetFont('helvetica', '', 12);

// Encabezado del reporte
$html = '<h1>Reporte de Empleados</h1>';

// Construir la tabla con los datos de la consulta SQL
$html .= '<table border="1">';
$html .= '<tr><th>ID</th><th>Nombre</th><th>Dirección</th><th>Salario</th></tr>';

if ($result->num_rows > 0) {
    while ($row = $result->fetch_assoc()) {
        $html .= '<tr>';
        $html .= '<td>' . $row['emp_id'] . '</td>';
        $html .= '<td>' . $row['emp_name'] . '</td>';
        $html .= '<td>' . $row['emp_address'] . '</td>';
        $html .= '<td>' . $row['emp_salary'] . '</td>';
        $html .= '</tr>';
    }
}

$html .= '</table>';

// Agregar el contenido HTML al PDF
$pdf->writeHTML($html, true, false, true, false, '');

// Cerrar la conexión
$conn->close();

// Salida del PDF al navegador
$pdf->Output('reporte_empleados.pdf', 'D');
?>

```
