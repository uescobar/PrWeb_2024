Las funciones agregadoras permiten agrupar la información para brindar un solo dato como salida de su aplicación, puede ser el conteo de algo, o bien, el promedio, moda, etc.

Apliquemos la primer función:

`SELECT COUNT(id), fecha 
    FROM reservaciones
    GROUP BY fecha
    ORDER BY COUNT(id) DESC;`

***
```
MariaDB [appsalon]> SELECT COUNT(id), fecha 
    -> FROM reservaciones
    -> GROUP BY fecha
    -> ORDER BY COUNT(id) DESC;
+-----------+------------+
| COUNT(id) | fecha      |
+-----------+------------+
|         7 | 2024-10-02 |
|         6 | 2024-10-01 |
|         6 | 2024-09-28 |
|         6 | 2024-10-30 |
|         6 | 2024-09-25 |
+-----------+------------+
5 rows in set (0.002 sec)

```
***

El día que hay más citas es en la fecha **2024-10-02**, la función agregadora es **COUNT**

Ahora se aplican a los servicios
