### INSTALACIÓN WAMPSERVER


Antes de realizar la instalación de WAMP server necesita los binarios de los redistribuibles de C++, desde 2012 hasta 2015 que puede descargar desde

[instalar esto primero, Redistribuibles Microsoft Visual C++ 2012, 2013, 2015](https://wampserver.aviatechno.net/#vcpackages)

Página desde donde los puede descargar, **debe instalar tanto la versión de 32 bits como la de 64 bits**, se encuentran señalados dentro del recuadro en color rojo los elementos que deben instalarse

![00](01_img/00.png)


Para el desarrollo de este módulo necesitaremos instalar los módulos de Apache, PHP, MySQL originalmente lo descargaremos para Windows, pero existe verison para Linux (aunque en linux lo mejor es instalar los servicios por separado para un entorno de producción).

Para comodidad y por simplicidad descargaremos el boundle que contiene todo, tambien existen otras soluciones que incorporan todo como son [laragon](https://laragon.org/download/), sin embargo, por ser una solución más familiar y que se ha usado en proyectos anteriores usaremos lo siguiente:

[Descarga de WAMP (Window Web development environment)](https://www.wampserver.com/en/)


La descarga anterior contiene Apache2, PHP, MySQL y un administrador de MySQL denominado PhpMyAdmin

Pagina principal de WAMP Server

![01](01_img/01.png)

La descarga es del servidor basado en 64 bits, en el momento de realizar este manual contiene las siguientes vesiones:

### Configurar las variables de entorno

Cerca del reloj del sistema, aparecerá el icono W, darle clic con el boton derecho y dar clic en la ruta:

**Tools > Command windows > Command windows PHP**

Con lo anterior se abre una cconsola en donde aparece el siguiente texto:

***
```
Microsoft Windows [Versión 10.0.19045.4046]
(c) Microsoft Corporation. Todos los derechos reservados.

c:\wamp64\bin\php\php8.2.13>
```
***

La última línea pertenece a la ruta en donde se encuentra el ejecutable de PHP, mismo que agregaremos a las variables de entorno en dos ubicaciones diferentes.

La **primera ubicación** es escribiendo en el buscador del menú inicio la palabra **variable** y seleccionar la opción **Editar las variables de entorno del sistema**

![16](01_img/16.png)

En la ventana que se abre hay que dar clic en el botón **Variables de Entorno**

![17](01_img/17.png)

En la ventana que se abre seleccionar la variable **Path** y dar clic en el botón **Editar**

![18](01_img/18.png)

En la ventana que se abre, dar clic en el botón **nuevo** y escribir **c:\wamp64\bin\php\php8.2.13** , que es el texto que aparece en la terminal, pero sin el simbolo **> (mayor que)**

![19](01_img/19.png)

Posteriormente, dar clic en **Aceptar** en todas las ventanas que se habian abierto

La **segunda ubicación**, de forma temporal, es escribiendo en el buscador del menú inicio la palabra **powershell** y seleccionar la opción **Windows PowerShell**

![20](01_img/20.png)

En la terminal, escribir **$env:PATH += ";c:\wamp64\bin\php\php8.2.13"** (note que la linea que va entre comillas simples, es la misma que en la ubicación anterior) y presionar **Enter**

![21](01_img/21.png)


Lo anterior tambien se puede realizar en la terminal de Visual Studio Code para poder ejecutar nuestros scripts desde la terminal de el IDE.

![22](01_img/22.png)
