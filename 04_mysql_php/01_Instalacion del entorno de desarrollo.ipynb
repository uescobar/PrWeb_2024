### INSTALACIÓN WAMPSERVER


Antes de realizar la instalación de WAMP server necesita los binarios de los redistribuibles de C++, desde 2012 hasta 2015 que puede descargar desde

[instalar esto primero, Redistribuibles Microsoft Visual C++ 2012, 2013, 2015](https://wampserver.aviatechno.net/#vcpackages)

Página desde donde los puede descargar, **debe instalar tanto la versión de 32 bits como la de 64 bits**, se encuentran señalados dentro del recuadro en color rojo los elementos que deben instalarse

![00](01_img/00.png)


Para el desarrollo de este módulo necesitaremos instalar los módulos de Apache, PHP, MySQL originalmente lo descargaremos para Windows, pero existe verison para Linux (aunque en linux lo mejor es instalar los servicios por separado para un entorno de producción).

Para comodidad y por simplicidad descargaremos el boundle que contiene todo, tambien existen otras soluciones que incorporan todo como son [laragon](https://laragon.org/download/), sin embargo, por ser una solución más familiar y que se ha usado en proyectos anteriores usaremos lo siguiente:

[Descarga de WAMP (Window Web development environment)](https://www.wampserver.com/en/)


La descarga anterior contiene Apache2, PHP, MySQL y un administrador de MySQL denominado PhpMyAdmin

Pagina principal de WAMP Server

![01](01_img/01.png)

La descarga es del servidor basado en 64 bits, en el momento de realizar este manual contiene las siguientes vesiones:

***
```
Wampserver 3.3.2 installeur complet 64 bit – Apache 2.4.58 – PHP 7.4.33/8.0.30/8.1.26/8.2.13/8.3.0 – MySQL 8.2.0 – MariaDB 11.2.2 – PhpMyAdmin 5.2.1 – Adminer 4.8.1 – PhpSysInfo 3.4.3.
```
***

Una vez descargado, el archivo se llama **wampserver3.3.2_x64.exe** y hayq ue dar doble clic en él.

Es importante mencionar que no debe estar instalado previamente nunguno de los programas que lo componen, ya que en caso de tener instalado previamente alguno de sus componentes, podríamos obtener algún comportamiento inusual e inestable.

Seguiremos los pasos de la instalación.

Elegimos el idioma

![02](01_img/02.png)



Aceptamos el acuerdo de licencia

![03](01_img/03.png)


Debemos dar clic en **next** unicamente si ya hemos instalado los redistribuibles de Visual C++. **Debe leer los prerequisitos para instalación para mas información**

![04](01_img/04.png)


Aceptamos el destino de instalación por defecto

![05](01_img/05.png)


Comienza la instalación

![06](01_img/06.png)


Seleccionamos **Si**, para elegir otro navegador por defecto que usará WAMP

![07](01_img/07.png)


Elegiremos Google Chrome, pero hay que especificar que se localiza en **C:\Program Files\Google\Chrome\Application\chrome.exe**, para el caso particular de mi computadora, en su caso, puede ser diferente, pero tiene que usar la interfaz para buscarlo, seleccionarlo y dar clic en **abrir**.

![08](01_img/08.png)

Seleccionamos **Si**, para elegir otro editor de texto por defecto que usará WAMP,

![09](01_img/09.png)

En nuestro caso elegiremos **Visual Studio Code**, pero hay que especificar que se localiza en **C:\Users\Lambda\AppData\Local\Programs\Microsoft VS Code\Code.exe**, , para el caso particular de mi computadora, en su caso, puede ser diferente, pero tiene que usar la interfaz para buscarlo, seleccionarlo y dar clic en **abrir**.

![10](01_img/10.png)



Revisar la información util que presenta y dar clic

![11](01_img/11.png)


La instalación terminada presenta la siguiente ventana a la que hay que dar clic en **Finalizar**

![12](01_img/12.png)


En el escritorio dar doble clic en el icono, otrorgando privilegios de administrador.

![13](01_img/13.png)


Cerca del reloj del sistema, aparecerá el icono **W**, darle clic y seleccionar **Start All Services**

![14](01_img/14.png)


Ahora, En el mismo Menu, seleccionar Localhost

![14](01_img/14.png)

Y el navegador debe desplegar la siguiente página web

![15](01_img/15.png)



**Una vez Visualizada la página anterior, podemos ver que PHP ha sido instalado correctamente, es probable que cada vez que iniciemos la máquina, tengamos que iniciar todos los servicios de esta manera**

Cerca del reloj del sistema, aparecerá el icono **W**, darle clic y seleccionar **Start All Services**
